# **DaÄŸÄ±lÄ±m Ã–lÃ§Ã¼tleri: Standart Sapma**ðŸŸ¢
**Standart sapma**, bir veri setindeki deÄŸerlerin ortalama deÄŸere ne kadar yayÄ±ldÄ±ÄŸÄ±nÄ± Ã¶lÃ§en bir istatistiksel Ã¶lÃ§Ã¼dÃ¼r. Yani, standart sapma, veri noktalarÄ±nÄ±n ortalama deÄŸerden ne kadar uzak olduÄŸunu gÃ¶sterir ve veri setinin daÄŸÄ±lÄ±mÄ±nÄ± ifade eder. Daha yÃ¼ksek bir standart sapma, veri noktalarÄ±nÄ±n ortalama deÄŸere daha fazla yayÄ±ldÄ±ÄŸÄ±nÄ± gÃ¶sterirken, daha dÃ¼ÅŸÃ¼k bir standart sapma, veri noktalarÄ±nÄ±n ortalama deÄŸere daha yakÄ±n olduÄŸunu gÃ¶sterir.

**PopÃ¼lasyon Standart SapmasÄ± FormÃ¼lÃ¼:
PopÃ¼lasyon standart sapmasÄ±, bir veri setindeki "N" adet deÄŸerin ortalama deÄŸerinden ne kadar uzak olduÄŸunun Ã¶lÃ§Ã¼sÃ¼dÃ¼r. PopÃ¼lasyon standart sapmasÄ± "Ïƒ" ile gÃ¶sterilir ve aÅŸaÄŸÄ±daki formÃ¼lle hesaplanÄ±r:

popÃ¼.PNG

**Ã–rnekleme GÃ¶re Standart Sapma FormÃ¼lÃ¼:**
Ã–rnekleme gÃ¶re standart sapma, bir veri setinden alÄ±nan Ã¶rneklemin ortalama deÄŸerinden ne kadar uzak olduÄŸunun Ã¶lÃ§Ã¼sÃ¼dÃ¼r. Ã–rnekleme gÃ¶re standart sapma "s" ile gÃ¶sterilir ve Bessel dÃ¼zeltmesi uygulanarak aÅŸaÄŸÄ±daki formÃ¼lle hesaplanÄ±r:

Ã¶rne.PNG

Bu formÃ¼ller, bir veri setinin standart sapmasÄ±nÄ± popÃ¼lasyon ve Ã¶rnekleme gÃ¶re hesaplamanÄ±n temelini oluÅŸturur.

```python
import numpy as np

# Rastgele veri seti oluÅŸturma
np.random.seed(0)  # Ã‡alÄ±ÅŸmanÄ±n tekrarlanabilir olmasÄ± iÃ§in random seed ayarlanÄ±yor
veri_seti = np.random.randint(0, 100, 100)  # 0 ile 100 arasÄ±nda rastgele 100 sayÄ±dan oluÅŸan bir veri seti


# PopÃ¼lasyon standart sapmasÄ± hesaplama
populasyon_std = np.std(veri_seti)

# Ã–rnekleme gÃ¶re standart sapma hesaplama
ornekleme_std = np.std(veri_seti, ddof=1)  # ddof=1, Bessel dÃ¼zeltmesini saÄŸlar

print("PopÃ¼lasyon Standart SapmasÄ±:", populasyon_std)
print("Ã–rnekleme GÃ¶re Standart Sapma:", ornekleme_std)
```

# **DaÄŸÄ±lÄ±m Ã–lÃ§Ã¼tleri: Varyans**ðŸŸ¢
**Varyans**, bir veri setindeki deÄŸerlerin ortalama deÄŸere ne kadar yayÄ±ldÄ±ÄŸÄ±nÄ± Ã¶lÃ§en bir istatistiksel Ã¶lÃ§Ã¼dÃ¼r. Yani, varyans, veri noktalarÄ±nÄ±n ortalama deÄŸerden ne kadar uzak olduÄŸunu gÃ¶sterir ve veri setinin daÄŸÄ±lÄ±mÄ±nÄ± ifade eder. Daha yÃ¼ksek bir varyans, veri noktalarÄ±nÄ±n ortalama deÄŸere daha fazla yayÄ±ldÄ±ÄŸÄ±nÄ± gÃ¶sterirken, daha dÃ¼ÅŸÃ¼k bir varyans, veri noktalarÄ±nÄ±n ortalama deÄŸere daha yakÄ±n olduÄŸunu gÃ¶sterir.

**PopÃ¼lasyon VaryansÄ± FormÃ¼lÃ¼:**
PopÃ¼lasyon varyansÄ±, bir veri setindeki "N" adet deÄŸerin ortalama deÄŸere ne kadar uzak olduÄŸunun karelerinin ortalamasÄ±dÄ±r. PopÃ¼lasyon varyansÄ± "Ïƒ^2" ile gÃ¶sterilir ve aÅŸaÄŸÄ±daki formÃ¼lle hesaplanÄ±r:

var popÃ¼.PNG

**Ã–rnekleme GÃ¶re Varyans FormÃ¼lÃ¼:**
Ã–rnekleme gÃ¶re varyans, bir veri setinden alÄ±nan Ã¶rneklemin ortalama deÄŸerinden ne kadar uzak olduÄŸunun karelerinin ortalamasÄ±dÄ±r. Ã–rnekleme gÃ¶re varyans "s^2" ile gÃ¶sterilir ve Bessel dÃ¼zeltmesi uygulanarak aÅŸaÄŸÄ±daki formÃ¼lle hesaplanÄ±r:

var Ã¶rne.PNG

Bu formÃ¼ller, bir veri setinin varyansÄ±nÄ± popÃ¼lasyon ve Ã¶rnekleme gÃ¶re hesaplamanÄ±n temelini oluÅŸturur.

```python
import numpy as np

# Ã–rnek veri seti
veri_seti = [5, 8, 10, 12, 6, 9, 11, 14, 7]

# PopÃ¼lasyon varyansÄ± hesaplama
populasyon_varyansi = np.var(veri_seti)

# Ã–rnekleme gÃ¶re varyans hesaplama
ornekleme_varyansi = np.var(veri_seti, ddof=1)  # ddof=1, Bessel dÃ¼zeltmesini saÄŸlar

print("PopÃ¼lasyon VaryansÄ±:", populasyon_varyansi)
print("Ã–rnekleme GÃ¶re Varyans:", ornekleme_varyansi)
```

# **DaÄŸÄ±lÄ±m Ã–lÃ§Ã¼tleri: Skewness (Ã‡arpÄ±klÄ±k)**ðŸŸ¢

Skewness (Ã‡arpÄ±klÄ±k), bir olasÄ±lÄ±k daÄŸÄ±lÄ±mÄ±nÄ±n veya veri setinin simetrisizliÄŸini Ã¶lÃ§en bir istatistiksel terimdir. Bir daÄŸÄ±lÄ±mÄ±n Ã§arpÄ±klÄ±ÄŸÄ±, daÄŸÄ±lÄ±mÄ±n merkezi eÄŸilimine (ortalama, medyan, mod gibi) gÃ¶re nasÄ±l eÄŸildiÄŸini gÃ¶sterir.

Bir veri setinin Ã§arpÄ±klÄ±ÄŸÄ± Ã¼Ã§ farklÄ± ÅŸekilde olabilir:

**1.Negatif Ã‡arpÄ±klÄ±k (Negative Skewness):** Bir veri seti negatif Ã§arpÄ±ksa, bu, veri setinin sola doÄŸru Ã§arpÄ±k olduÄŸunu gÃ¶sterir. Yani, veri seti genellikle saÄŸ kuyruÄŸu uzun olan ve sola doÄŸru uzanan bir daÄŸÄ±lÄ±m sergiler. Medyan, ortalama deÄŸerden daha kÃ¼Ã§Ã¼k olacaktÄ±r.

**2.Pozitif Ã‡arpÄ±klÄ±k (Positive Skewness):** Bir veri seti pozitif Ã§arpÄ±ksa, bu, veri setinin saÄŸa doÄŸru Ã§arpÄ±k olduÄŸunu gÃ¶sterir. Yani, veri seti genellikle sol kuyruÄŸu uzun olan ve saÄŸa doÄŸru uzanan bir daÄŸÄ±lÄ±m sergiler. Medyan, ortalama deÄŸerden daha bÃ¼yÃ¼k olacaktÄ±r.

**3.SÄ±fÄ±r Ã‡arpÄ±klÄ±k (Zero Skewness):** Bir veri seti sÄ±fÄ±r Ã§arpÄ±ksa, bu, veri setinin simetrik olduÄŸunu ve saÄŸa veya sola doÄŸru bir eÄŸilimin olmadÄ±ÄŸÄ±nÄ± gÃ¶sterir. Bu durumda, veri setinin medyanÄ±, ortalama deÄŸere eÅŸit olacaktÄ±r.

Ã‡arpÄ±klÄ±k, veri setinin daÄŸÄ±lÄ±mÄ±nÄ±n gÃ¶rselleÅŸtirilmesi ve anlaÅŸÄ±lmasÄ± iÃ§in Ã¶nemlidir. Ã–zellikle, veri setinin simetrik olup olmadÄ±ÄŸÄ±nÄ± ve merkezi eÄŸilim Ã¶lÃ§Ã¼leriyle (Ã¶rneÄŸin, ortalama ve medyan) nasÄ±l iliÅŸkili olduÄŸunu belirlemek iÃ§in kullanÄ±lÄ±r. Bir veri setinin Ã§arpÄ±klÄ±ÄŸÄ±nÄ± anlamak, veri setinin doÄŸru analiz edilmesi ve yorumlanmasÄ± iÃ§in Ã¶nemlidir.

**Ã–rnek 1:**
Bu kod Ã¶rneÄŸinde, Ã¶ncelikle NumPy kÃ¼tÃ¼phanesini kullanarak bir normal daÄŸÄ±lÄ±ma sahip 1000 elemanlÄ± bir Ã¶rnek veri seti oluÅŸturuyoruz. Daha sonra, oluÅŸturduÄŸumuz bu veri setinin histogramÄ±nÄ± Ã§iziyoruz. HistogramÄ± Ã§izerken, plt.hist() fonksiyonunu kullanarak veri setini, istenen bin sayÄ±sÄ± ve diÄŸer grafik ayarlarÄ±nÄ± belirtiyoruz.

ArdÄ±ndan, scipy.stats.skew() fonksiyonunu kullanarak veri setinin Ã§arpÄ±klÄ±ÄŸÄ±nÄ± hesaplÄ±yoruz. Elde ettiÄŸimiz Ã§arpÄ±klÄ±k deÄŸerini, histogram grafiÄŸinin Ã¼zerine yazdÄ±rÄ±yoruz.

Son olarak, Ã§izilen grafik ve hesaplanan Ã§arpÄ±klÄ±k deÄŸeri ile birlikte bir grafik oluÅŸturuyoruz. Bu ÅŸekilde, veri setinin daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶rselleÅŸtiriyoruz ve Ã§arpÄ±klÄ±k hakkÄ±nda bilgi elde ediyoruz.

```python
#Ã–rnek 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Ã–rnek veri seti oluÅŸturma
np.random.seed(0)
veri_seti = np.random.normal(loc=10, scale=5, size=1000)  # OrtalamasÄ± 10, standart sapmasÄ± 5 olan 1000 elemanlÄ± normal daÄŸÄ±lÄ±m

# Veri setinin histogramÄ±nÄ± Ã§izme
plt.hist(veri_seti, bins=30, density=True, alpha=0.6, color='g')

# Veri setinin Ã§arpÄ±klÄ±ÄŸÄ±nÄ± hesaplama
veri_seti_carpiklik = skew(veri_seti)

# Ã‡arpÄ±klÄ±k deÄŸerini ekrana yazdÄ±rma
plt.text(30, 0.05, f"Ã‡arpÄ±klÄ±k: {veri_seti_carpiklik:.2f}", fontsize=15)

# Grafik ayarlarÄ±
plt.xlabel('DeÄŸerler')
plt.ylabel('YoÄŸunluk')
plt.title('Veri Seti HistogramÄ± ve Ã‡arpÄ±klÄ±k')
plt.grid(True)
plt.show()
```

**Ã–rnek 2:**
Bu kodda, Ã¶nce farklÄ± daÄŸÄ±lÄ±mlara sahip Ã¼Ã§ Ã¶rnek veri seti oluÅŸturuyoruz: normal, gamma ve log-normal daÄŸÄ±lÄ±mlarÄ±. Daha sonra, her bir veri setinin Ã§arpÄ±klÄ±ÄŸÄ±nÄ± hesaplayÄ±p, her biri iÃ§in histogramlarÄ±nÄ± Ã§izdiriyoruz. Her histogramÄ±n baÅŸlÄ±ÄŸÄ±nda, hesaplanan Ã§arpÄ±klÄ±k deÄŸeri yer alÄ±yor. Bu ÅŸekilde, farklÄ± daÄŸÄ±lÄ±mlara sahip veri setlerinin Ã§arpÄ±klÄ±klarÄ±nÄ± gÃ¶rselleÅŸtirip, karÅŸÄ±laÅŸtÄ±rma yapabiliyoruz.

```python
#Ã–rnek 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, gamma, lognorm

# Ã–rnek veri seti oluÅŸturma
np.random.seed(0)
veri_seti_normal = np.random.normal(loc=0, scale=1, size=1000)  # Normal daÄŸÄ±lÄ±m
veri_seti_gamma = gamma.rvs(a=2, size=1000)  # Gamma daÄŸÄ±lÄ±mÄ±
veri_seti_lognorm = lognorm.rvs(s=0.5, size=1000)  # Log-normal daÄŸÄ±lÄ±mÄ±

# Veri setlerinin Ã§arpÄ±klÄ±ÄŸÄ±nÄ± hesaplama
normal_carpiklik = skew(veri_seti_normal)
gamma_carpiklik = skew(veri_seti_gamma)
lognorm_carpiklik = skew(veri_seti_lognorm)

# HistogramlarÄ± Ã§izdirme
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(veri_seti_normal, bins=30, density=True, alpha=0.6, color='g')
plt.title(f'Normal DaÄŸÄ±lÄ±m\nÃ‡arpÄ±klÄ±k: {normal_carpiklik:.2f}')

plt.subplot(1, 3, 2)
plt.hist(veri_seti_gamma, bins=30, density=True, alpha=0.6, color='b')
plt.title(f'Gamma DaÄŸÄ±lÄ±mÄ±\nÃ‡arpÄ±klÄ±k: {gamma_carpiklik:.2f}')

plt.subplot(1, 3, 3)
plt.hist(veri_seti_lognorm, bins=30, density=True, alpha=0.6, color='r')
plt.title(f'Log-Normal DaÄŸÄ±lÄ±mÄ±\nÃ‡arpÄ±klÄ±k: {lognorm_carpiklik:.2f}')

plt.tight_layout()
plt.show()
```

# **DaÄŸÄ±lÄ±m Ã–lÃ§Ã¼tleri: Kurtosis (BasÄ±klÄ±k)**ðŸŸ¢

**Kurtosis**, bir olasÄ±lÄ±k daÄŸÄ±lÄ±mÄ±nÄ±n veya veri setinin sivriÄŸi veya dÃ¼zleÄŸi hakkÄ±nda bilgi veren bir istatistiksel terimdir. Yani, bir daÄŸÄ±lÄ±mÄ±n kurtosis deÄŸeri, daÄŸÄ±lÄ±mÄ±n zirvesinin (tepe noktasÄ±nÄ±n) ne kadar sivri veya dÃ¼z olduÄŸunu Ã¶lÃ§er. Kurtosis, daÄŸÄ±lÄ±mÄ±n uÃ§ deÄŸerlerinin yoÄŸunlaÅŸmasÄ±nÄ± inceler.

Kurtosis deÄŸeri, genellikle dÃ¶rt farklÄ± tÃ¼rde tanÄ±mlanÄ±r:

**1. Mesokurtik (Mesokurtic):** Normal daÄŸÄ±lÄ±ma (standart normal daÄŸÄ±lÄ±ma) benzer bir kurtosis yapÄ±sÄ±na sahip daÄŸÄ±lÄ±mlar mesokurtik olarak adlandÄ±rÄ±lÄ±r. Mesokurtik bir daÄŸÄ±lÄ±m, ortalama kurtosis deÄŸeri olan 0'a sahiptir.

**2. Leptokurtik (Leptokurtic):** Leptokurtik bir daÄŸÄ±lÄ±m, normal daÄŸÄ±lÄ±ma gÃ¶re daha sivri ve daha az yayÄ±lmÄ±ÅŸtÄ±r. UÃ§ deÄŸerlerin normal daÄŸÄ±lÄ±ma gÃ¶re daha yoÄŸun olduÄŸu durumlarda leptokurtik bir kurtosis deÄŸeri elde edilir. Leptokurtik bir daÄŸÄ±lÄ±m, kurtosis deÄŸeri pozitif olan bir daÄŸÄ±lÄ±mdÄ±r.

**3. Platykurtik (Platykurtic):** Platykurtik bir daÄŸÄ±lÄ±m, normal daÄŸÄ±lÄ±ma gÃ¶re daha dÃ¼z ve daha geniÅŸ bir yayÄ±lÄ±ma sahiptir. UÃ§ deÄŸerlerin normal daÄŸÄ±lÄ±ma gÃ¶re daha az yoÄŸun olduÄŸu durumlarda platykurtik bir kurtosis deÄŸeri elde edilir. Platykurtik bir daÄŸÄ±lÄ±m, kurtosis deÄŸeri negatif olan bir daÄŸÄ±lÄ±mdÄ±r.


*Ã–zetle, kurtosis deÄŸeri, bir veri setinin daÄŸÄ±lÄ±mÄ±nÄ±n sivriÄŸi veya dÃ¼zleÄŸi hakkÄ±nda bilgi saÄŸlar. Mesokurtik daÄŸÄ±lÄ±mlar genellikle simetrik ve normal daÄŸÄ±lÄ±ma benzerken, leptokurtik ve platykurtik daÄŸÄ±lÄ±mlar daha farklÄ± Ã¶zelliklere sahiptir.*

Bu programda, farklÄ± daÄŸÄ±lÄ±mlara sahip Ã¼Ã§ Ã¶rnek veri setinin kurtosis deÄŸerlerini hesaplayacaÄŸÄ±z ve bunlarÄ± karÅŸÄ±laÅŸtÄ±rmak iÃ§in bir grafik oluÅŸturacaÄŸÄ±z. Ä°ÅŸte bu program iÃ§in bir Python kodu:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, gamma, lognorm

# Ã–rnek veri setlerini oluÅŸturma
np.random.seed(0)
normal_data = np.random.normal(loc=0, scale=1, size=1000)  # Normal daÄŸÄ±lÄ±m
gamma_data = gamma.rvs(a=2, size=1000)  # Gamma daÄŸÄ±lÄ±mÄ±
lognorm_data = lognorm.rvs(s=0.5, size=1000)  # Log-normal daÄŸÄ±lÄ±mÄ±

# Veri setlerinin kurtosis deÄŸerlerini hesaplama
kurtosis_normal = kurtosis(normal_data)
kurtosis_gamma = kurtosis(gamma_data)
kurtosis_lognorm = kurtosis(lognorm_data)

# HistogramlarÄ± Ã§izdirme
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='g')
plt.title(f'Normal DaÄŸÄ±lÄ±m\nKurtosis: {kurtosis_normal:.2f}')

plt.subplot(1, 3, 2)
plt.hist(gamma_data, bins=30, density=True, alpha=0.6, color='b')
plt.title(f'Gamma DaÄŸÄ±lÄ±mÄ±\nKurtosis: {kurtosis_gamma:.2f}')

plt.subplot(1, 3, 3)
plt.hist(lognorm_data, bins=30, density=True, alpha=0.6, color='r')
plt.title(f'Log-Normal DaÄŸÄ±lÄ±mÄ±\nKurtosis: {kurtosis_lognorm:.2f}')

plt.tight_layout()
plt.show()
```

Bu kod, normal, gamma ve log-normal daÄŸÄ±lÄ±mlarÄ± iÃ§in Ã¶rnek veri setleri oluÅŸturur. Daha sonra, her bir veri setinin kurtosis deÄŸerini hesaplar ve bu deÄŸerleri histogramlarÄ±n baÅŸlÄ±klarÄ±nda gÃ¶sterir. SonuÃ§ olarak, Ã¼Ã§ farklÄ± daÄŸÄ±lÄ±mÄ±n kurtosis deÄŸerlerini karÅŸÄ±laÅŸtÄ±ran bir grafik elde ederiz.
